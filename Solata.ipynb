```python
Solata_vsi_podatki['Date Time'] = pd.to_datetime(Solata_vsi_podatki['Date Time'], format = '%d.%m.%Y %H:%M')
Solata_vsi_podatki = Solata_vsi_podatki.set_index('Date Time', drop=True)
Solata_vsi_podatki = Solata_vsi_podatki.sort_index()
Solata_vsi_podatki
```

Združevanje po sondah.

Dimenzije tabele

```python
st_stolpcev = len(solata_po_casu.columns)
st_vrstic = len(solata_po_casu)
```

### Računanje razlik, hitogrami in ___ grafi. 

```python
for j in range(st_stolpcev):

    razlike = []
    trenutna_vrednost = -99
    stevec_NaN = 1
    for i in range(st_vrstic):
        if solata_po_casu['Vlaga{}'.format(j+1)].isnull()[i]:
            if trenutna_vrednost == -99:
                razlike.append(np.nan)
            else:
                stevec_NaN +=1
                razlike.append(np.nan)
        else:
            if trenutna_vrednost == -99:
                razlike.append(0)
                
            else:
                razlike.append((solata_po_casu['Vlaga{}'.format(j+1)][i]-trenutna_vrednost)/stevec_NaN)
                stevec_NaN = 1
            trenutna_vrednost = solata_po_casu['Vlaga{}'.format(j+1)][i]
    
    stolpci = {'Cas':solata_po_casu.index,'vrednosti': solata_po_casu['Vlaga{}'.format(j+1)], 'razlike': razlike} 
    vrednosti_razlike = pd.DataFrame(data=stolpci)
    vrednosti_razlike.plot(secondary_y=['razlike']).set_ylabel('vrednosti')
    
    fig, axes = plt.subplots(1,2)
    solata_po_casu['Vlaga{}'.format(j+1)].hist(ax=axes[0])
    pd.Series(razlike).hist(ax=axes[1])
    axes[0].set_title('Vrednosti')
    axes[1].set_title('Razlike')

```

```python
stolpci ={}

for j in range(st_stolpcev):
    razlike = []
    trenutna_vrednost = -99
    stevec_NaN = 1
    for i in range(st_vrstic):
        if solata_po_casu['Vlaga{}'.format(j+1)].isnull()[i]:
            if trenutna_vrednost == -99:
                razlike.append(np.nan)
            else:
                stevec_NaN +=1
                razlike.append(np.nan)
        else:
            if trenutna_vrednost == -99:
                razlike.append(0)
                
            else:
                razlike.append((solata_po_casu['Vlaga{}'.format(j+1)][i]-trenutna_vrednost)/stevec_NaN)
                stevec_NaN = 1
            trenutna_vrednost = solata_po_casu['Vlaga{}'.format(j+1)][i]
    stolpci['Vlaga{}'.format(j+1)] =solata_po_casu['Vlaga{}'.format(j+1)]
    stolpci['razlike{}'.format(j+1)] =razlike
    #print(stolpci)
vrednosti_razlike= pd.DataFrame(stolpci)

```
